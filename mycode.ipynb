* volkswagen is the most popular car in germen

```python
top20_brands = autos.brand.value_counts().head(20).index
```

```python
top20_brands_mean_price = {}
for item in top20_brands:
    selected_rows = autos[autos.brand == item]
    v = round(selected_rows.price.mean())
    top20_brands_mean_price[item] = v
```

```python
top20_brands_mean_price
```

* sonstige_autos has the highest average price

```python
top20_brands_mean_price = {}
top20_brands_mean_mileage = {}
for item in top20_brands:
    selected_rows = autos[autos.brand == item]
    mean_price = round(selected_rows.price.mean())
    mean_mileage = round(selected_rows.odometer_km.mean())
    top20_brands_mean_price[item] = mean_price
    top20_brands_mean_mileage[item] = mean_mileage
```

```python
top20_brands_mean_price = pd.Series(top20_brands_mean_price)
top20_brands_mean_mileage = pd.Series(top20_brands_mean_mileage)
```

```python
top20_brands_mean_price = pd.DataFrame(top20_brands_mean_price,columns = ['mean_price'])
```

```python
top20_brands_mean_price['mean_mileage'] = top20_brands_mean_mileage
```

```python
top20_branks_mean_price_mileage = top20_brands_mean_price
```

```python
top20_branks_mean_price_mileage.sort_values(by='mean_price')
```

* audi, mercedes_benz, and bmw have high price even though they have high mileage
